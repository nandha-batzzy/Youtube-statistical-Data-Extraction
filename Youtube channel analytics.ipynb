```python
import json 
import pandas as pd
from datetime  import datetime
import matplotlib.pyplot as plt
with open('think_beyond_-_tamil.json','r') as f:
    data = json.load(f)
chann_id,stats = data.popitem()
stats = stats['video_data']
sorted_stats = sorted(stats.items(), key= lambda item: int(item[1]['viewCount']), reverse = True)
stats1 = []
for stats in sorted_stats:
    video_id = stats[0]
    aired_at = stats[1]['publishedAt']
    title = stats[1]['title']
    view = stats[1]['viewCount']
    likes = stats[1]['likeCount']
    dislike = stats[1]['dislikeCount']
    comments = stats[1]['commentCount']
    duration = stats[1]['duration']
    stats1.append([aired_at,title,view,likes,dislike,comments,duration])
```

```python
df = pd.DataFrame(stats1,columns=['aired_at','title','view','likes','dislike','comments','duration'])
df
```

```python
df['aired_at']= pd.to_datetime(df['aired_at'])
```

```python
df = df.sort_values(by='aired_at',axis=0)
```

```python
df['view'] = pd.to_numeric(df['view'])
df['likes'] = pd.to_numeric(df['likes'])
df['dislike'] = pd.to_numeric(df['dislike'])
df['comments'] = pd.to_numeric(df['comments'])
```

```python
import seaborn as sns
```

```python
mdf = df.sort_values(by='view',axis=0,ascending=False)
top_10 = mdf.head(10)
top_10
```

```python
f, ax = plt.subplots(nrows=1, ncols=1, figsize=(25, 10))
sns.barplot(x = top_10['title'],y=top_10['view'],ax=ax)
plt.xticks(rotation=90)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 18)
```

```python
df['no_dislikes_per_like'] = df['dislike']/df['likes']
```

```python
mdf = df.sort_values(by='no_dislikes_per_like',axis=0,ascending=False)
top_10 = mdf.head(10)
top_10
```

```python
f, ax = plt.subplots(nrows=1, ncols=1, figsize=(25, 10))
sns.barplot(x = top_10['title'],y=top_10['no_dislikes_per_like'],ax=ax)
plt.xticks(rotation=90)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 18)
```

```python
df['likes_per_view'] = df['likes']/df['view']
mdf = df.sort_values(by='likes_per_view',axis=0,ascending=False)
top_10 = mdf.head(10)
top_10
```

```python
f, ax = plt.subplots(nrows=1, ncols=1, figsize=(25, 10))
sns.barplot(x = top_10['title'],y=top_10['likes_per_view'],ax=ax)
plt.xticks(rotation=90)
ax.set_xticklabels(ax.get_xmajorticklabels(), fontsize = 18)
```
